# generating a list of every sunday between today and 15 years ago

```python
import datetime

sunday_dates = []

current_date = datetime.date.today() # current date operating on 5-17-2023

start_date = current_date - datetime.timedelta(days=15 * 365)

current = start_date
while current <= current_date:
    if current.weekday() == 6:
        print(current)
        sunday_dates.append(current.strftime("%Y-%m-%d"))
    current += datetime.timedelta(days=1)
```

# TEST RUN - testing dates that pre-dated the bestseller list + age groups that were not on the list for specific dates to see what happens

```python
second_test = ['2008-10-05','2022-08-14', '2008-05-25', '2022-08-21']
new_test = ['hardcover-fiction', 'young-adult-paperback-monthly', 'indigenous-americans']
new_data = []

for date in second_test:
    for new in new_test:
        print(f"Currently Processesing: {date} {new}")
        url2 = f"https://api.nytimes.com/svc/books/v3/lists/{date}/{new}.json?api-key={APIKEY}"
        time.sleep(12)
        history_response2 = requests.get(url2)
        if history_response2.status_code == 200:
            history_test2 = history_response2.json()
            new_data.append(history_test2)
```

# IT'S SHOW TIME

```python
data_2008 = []

for date in year_2008:
    print(f"Currently Processesing: {date}")
    url = f"https://api.nytimes.com/svc/books/v3/lists/full-overview.json?api-key={APIKEY}"
    time.sleep(12)
    my_parameters = {'published_date' : date}
    response_2008 = requests.get(url, params=my_parameters)
    if response_2008.status_code == 200:
        data = response_2008.json()
        data_2008.append(data)
```

```python
with open('2008.json', 'w') as file:
    json.dump(data_2008, file)
```
