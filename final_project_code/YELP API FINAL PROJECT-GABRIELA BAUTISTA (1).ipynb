```python
import yelp_authentication;
import requests;
import json;
import yelpapi;

def do_yelp_request(term, location, all=0):
    headers = {
        'Authorization': 'Bearer ' + yelp_authentication.yelp_api_key
    }

    url = 'https://api.yelp.com/v3/businesses/search'
    limit = 50
    radius = 1000
    #offset = 0
    params = {
        'location': location,
        'term': term,
        'limit': limit,
        'radius':radius
        #'offset': offset
    }
    
    all_results = []  # List to store all results

    while True:
    #for offset in range(0, 1000, 50):
        print("Requesting yelp...")
        #print(offset)
        response = requests.get(url, headers=headers, params=params)
        
        print("ratelimit-remaining:"+response.headers['ratelimit-remaining'])
        data = response.json()
        if 'businesses' in data:
            data["businesses"]
            all_results.extend(data['businesses'])

        total_results = data['total']
        fetched_results = len(all_results)

        if (all == 0):
            break

        if fetched_results < total_results:
            # Increment the offset to fetch the next page
            params['offset'] += limit
        else:
            # No more results to fetch, exit the loop
            break
        
    return all_results
```

```python
#res = do_yelp_request("restaurants","98122")
#Cap hill

capitol_hill_zipcode = ["98112", "98122","98122"]

def restaurants_byzip(zip_code_list):
    restaurants_in_zipcode = []
    for zipcode in zip_code_list:
        restaurants_in_zipcode.append(do_yelp_request("restaurants",zipcode))
        
    return restaurants_in_zipcode

caphill_restaurants = restaurants_byzip(capitol_hill_zipcode)
```

```python
def get_categories(input_res):
    #print(input_res)
    list_categories = {}

    for zip_code_list in input_res:
        for item in zip_code_list:
            item_categories = item['categories']
            for category in item_categories:
                category_alias = category['alias']
                if category_alias in list_categories:
                    list_categories[category_alias]+=1
                else:
                    list_categories[category_alias]=1
    return list_categories

cap_hill_categories = get_categories(caphill_restaurants)
print(cap_hill_categories)
```
