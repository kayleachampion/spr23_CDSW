```python
import json
```

```python
#Opening the JSON file downloaded through Data Collecting script

with open("india_rainfall_data_from_2009_to_2016.json", 'r') as input_file:
    input_data = input_file.read()
    rainfall_data = json.loads(input_data)
```

```python
#The data is on rainfall in individual districts. I would like to get it by state. Starting with just year 2009.

import pandas as pd

# Create an empty DataFrame to store the results
result_df = pd.DataFrame(columns=['Date', 'Indian_States_Name', 'Total_Value'])

# Iterate over each item in the rainfall_data list
for item in rainfall_data:
    fields = item['fields']
    date = fields['date']
    year = date.split('-')[0]  # Extract the year from the date
    
    # Check if the year is 2009
    if year == '2009':
        indian_states_name = fields['indian_states_name']
        value = fields['value']
        
        # Check if the Indian state name and date combination already exists in the result DataFrame
        matching_rows = result_df[(result_df['Date'] == date) & (result_df['Indian_States_Name'] == indian_states_name)]
        
        if not matching_rows.empty:
            # If it exists, update the total value by adding the current value
            result_df.loc[matching_rows.index, 'Total_Value'] += value
        else:
            # If it doesn't exist, create a new row with the Indian state name, date, and value
            new_row = pd.DataFrame({'Date': [date], 'Indian_States_Name': [indian_states_name], 'Total_Value': [value]})
            result_df = pd.concat([result_df, new_row], ignore_index=True)

# Print the resulting DataFrame
print(result_df)
```

```python
import pandas as pd

# Create an empty DataFrame to store the results
result_df = pd.DataFrame(columns=['Year', 'Indian_States_Name', 'Total_Value'])

# Iterate over each item in the rainfall_data list
for item in rainfall_data:
    fields = item['fields']
    date = fields['date']
    year = date.split('-')[0]  # Extract the year from the date
    
    # Check if the year is 2009
    if year == '2009':
        indian_states_name = fields['indian_states_name']
        value = fields['value']
        
        # Check if the Indian state name and year combination already exists in the result DataFrame
        matching_rows = result_df[(result_df['Year'] == year) & (result_df['Indian_States_Name'] == indian_states_name)]
        
        if not matching_rows.empty:
            # If it exists, update the total value by adding the current value
            result_df.loc[matching_rows.index, 'Total_Value'] += value
        else:
            # If it doesn't exist, create a new row with the Indian state name, year, and value
            new_row = pd.DataFrame({'Year': [year], 'Indian_States_Name': [indian_states_name], 'Total_Value': [value]})
            result_df = pd.concat([result_df, new_row], ignore_index=True)

# Group the result DataFrame by year and state, and calculate the sum of values within each group
result_df = result_df.groupby(['Year', 'Indian_States_Name']).sum().reset_index()

# Print the resulting DataFrame
print(result_df)
```

```python
#Now Collecting data for years from 2009 to 2015.

import pandas as pd

# Create an empty DataFrame to store the results
result_df = pd.DataFrame(columns=['Year', 'Indian_States_Name', 'Total_Value'])

# Iterate over each item in the rainfall_data list
for item in rainfall_data:
    fields = item['fields']
    date = fields['date']
    year = int(date.split('-')[0])  # Extract the year from the date
    
    # Check if the year is between 2009 and 2015 (inclusive)
    if 2009 <= year <= 2015:
        indian_states_name = fields['indian_states_name']
        value = fields['value']
        
        # Check if the Indian state name and year combination already exists in the result DataFrame
        matching_rows = result_df[(result_df['Year'] == year) & (result_df['Indian_States_Name'] == indian_states_name)]
        
        if not matching_rows.empty:
            # If it exists, update the total value by adding the current value
            result_df.loc[matching_rows.index, 'Total_Value'] += value
        else:
            # If it doesn't exist, create a new row with the Indian state name, year, and value
            new_row = pd.DataFrame({'Year': [year], 'Indian_States_Name': [indian_states_name], 'Total_Value': [value]})
            result_df = pd.concat([result_df, new_row], ignore_index=True)

# Group the result DataFrame by year and state, and calculate the sum of values within each group
result_df = result_df.groupby(['Year', 'Indian_States_Name']).sum().reset_index()

# Sort the DataFrame by year in ascending order
result_df = result_df.sort_values('Year')

# Print the resulting DataFrame
print(result_df)
```

```python
# Pivot the result DataFrame to have years as columns
result_pivot = result_df.pivot(index='Indian_States_Name', columns='Year', values='Total_Value').reset_index()

# Rename the column names to include the 'total value' prefix
result_pivot.columns = ['Indian_States_Name'] + ['Total_Value_' + str(year) for year in result_pivot.columns[1:]]

# Print the resulting DataFrame
print(result_pivot)
```

```python
result_pivot.to_csv('indian_states_rainfall_data_by_year.tsv', sep='\t', index=False)
```
