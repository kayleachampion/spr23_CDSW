```python
# Collecting data on Rainfall in India for year 2009 to check out the data:

url = "https://datasource.kapsarc.org/api/records/1.0/search/"
params = {
    "dataset": "district-wise-rainfall-data-for-india-2014",
    "q": "",
    "rows": 10000,
    "sort": "-date",
    "facet": ["indian_states_name", "districts_name", "date", "variable_name"],
    "refine.variable_name": "Rainfall (mm.)",
    "refine.date": 2009
}

response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print("Error occurred. Status Code:", response.status_code)
```

```python
# Collecting data for years 2009 to 2015

import requests
import json

base_url = "https://datasource.kapsarc.org/api/records/1.0/search/"
dataset = "district-wise-rainfall-data-for-india-2014"
refine_variable_name = "Rainfall (mm.)"

years = [2009,2010,2011,2012,2013,2014,2015]
all_data = []

for year in years:
    params = {
        "dataset": dataset,
        "q": "",
        "rows": 10000,
        "sort": "-date",
        "facet": ["indian_states_name", "districts_name", "date", "variable_name"],
        "refine.variable_name": refine_variable_name,
        "refine.date": year
    }

    response = requests.get(base_url, params=params)

    if response.status_code == 200:
        data = response.json()
        all_data.extend(data["records"])
        print(f"Data for {year} retrieved successfully.")
    else:
        print(f"Error occurred for {year}. Status Code:", response.status_code)

# Save all data in a JSON file
file_name = "india_rainfall_data_from_2009_to_2016.json"
with open(file_name, "w") as json_file:
    json.dump(all_data, json_file)

print(f"All data saved successfully in {file_name}.")
```
