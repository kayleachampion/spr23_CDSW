# word cloud analysis and count

```python
import json
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

file_paths = ['may24_info.json', 'may25_info.json', 'may26_info.json', 'may27_info.json', 'may28_info.json']

all_tags = []
for path in file_paths:
    with open(path, 'r') as file:
        data = json.load(file)
        for item in data:
            tags = item.get('tags')
            if tags is not None:
                all_tags.extend(tags)

wordcloud = WordCloud(width=800, height=400, max_words=200).generate(' '.join(all_tags))

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import pandas as pd
import re

file_paths = ['may24_info.json', 'may25_info.json', 'may26_info.json', 'may27_info.json', 'may28_info.json']
titles_languages = []
titles_question_mark = []
titles_incomplete_sentences = []
titles_caps_lock = []

for file_path in file_paths:
    with open(file_path, 'r') as file:
        data = json.load(file)
        df = pd.DataFrame(data)  # Create a pandas DataFrame from the data
        titles = df['title']
        
        for title in titles:
            if not re.match(r'^[a-zA-Z0-9\s\.,\?!]+$', title):
                titles_languages.append(title)
            if '?' in title:
                titles_question_mark.append(title)
            if '.' in title:
                sentences = title.split('.')
                for sentence in sentences:
                    if sentence.strip() and not sentence.endswith('!') and not sentence.endswith('?'):
                        titles_incomplete_sentences.append(title)
                        break
            if title.isupper():
                titles_caps_lock.append(title)

print("Titles in other languages:")
print(titles_languages)
print("\nTitles with question marks:")
print(titles_question_mark)
print("\nTitles with incomplete sentences:")
print(titles_incomplete_sentences)
print("\nTitles in all caps:")
print(titles_caps_lock)
```

```python
import pandas as pd
import re
import langid
import matplotlib.pyplot as plt
import seaborn as sns

file_paths = ['may24_info.json', 'may25_info.json', 'may26_info.json', 'may27_info.json', 'may28_info.json']
titles_languages = []
titles_question_mark = []
titles_incomplete_sentences = []
titles_caps_lock = []

for file_path in file_paths:
    with open(file_path, 'r') as file:
        data = json.load(file)
        df = pd.DataFrame(data)  # Create a pandas DataFrame from the data
        titles = df['title']
        
        for title in titles:
            if '?' in title:
                titles_question_mark.append(title)
            if '.' in title:
                sentences = title.split('.')
                for sentence in sentences:
                    if sentence.strip() and not sentence.endswith('!') and not sentence.endswith('?'):
                        titles_incomplete_sentences.append(title)
                        break
            if title.isupper():
                titles_caps_lock.append(title)

# Count the number of titles in each category
categories = ['Titles with Question Marks', 'Titles with Incomplete Sentences', 'Titles in All Caps']
counts = [len(titles_question_mark), len(titles_incomplete_sentences), len(titles_caps_lock)]

# Create a bar plot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=categories, y=counts)
ax.set_ylabel('Count')
ax.set_title('Title Categories')

# Add count labels on top of each bar
for i, count in enumerate(counts):
    ax.text(i, count + 10, str(count), ha='center')

plt.xticks(rotation=45)
plt.show()
```
