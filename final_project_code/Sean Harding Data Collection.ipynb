```python
import requests
import json
import time
```

```python
with open("output_1.jsonl", "w") as my_file:
    article_counter = 0
    print(article_counter)
    while article_counter < 150000:
        try:
            my_response = requests.get(f"https://api.army.mil/api/v1/article/{article_counter}")
            if my_response.status_code == 200:
                new_data = my_response.json()
                print(new_data, file = my_file)
                article_counter += 1
            else:
                print(f"{article_counter}, {my_response.status_code}")
                article_counter += 1
            
if article_counter == 150000:
    print("Data collection complete")
    print(article_counter)
else:
    print("Error")
    print(f"{article_counter}, {my_response.status_code}")
    #print(article_counter)
```

```python
with open("output_1.jsonl", "w") as my_file:
    article_counter = 0
    print(article_counter)
    while article_counter < 150000:
        try:
            my_response = requests.get(f"https://api.army.mil/api/v1/article/{article_counter}")
            if my_response.status_code == 200:
                new_data = my_response.json()
                new_data_string = json.dumps(new_data)
                print(new_data_string, file = my_file)
                article_counter += 1
            else:
                print(f"{article_counter}, {my_response.status_code}")
                article_counter += 1
        except (ConnectionError, NameError, OSError):
            print("Connection/Protocol error")
            time.sleep(60)
            continue

if article_counter == 150000:
    print("Data collection complete")
    print(article_counter)
else:
    print("Error")
    print(f"{article_counter}, {my_response.status_code}")
    #print(article_counter)
```
