# DATA COLLECTION - testing the openlibrary api

```python
import requests
import json
```

```python
def get_book_data(isbn): 
    
    url = f"https://openlibrary.org/isbn/{isbn}.json"
    
    headers = {
        'User-Agent' : 'python data collection bot by dtayl23@uw.edu'
    }

    response = requests.get(url, headers=headers)

    if not response.status_code == 200:
        print("ERROR")
        
    data = response.json()
    return data
```
