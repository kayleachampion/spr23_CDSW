# Category

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# File paths for the data
file_paths = ['categorized_dataMay24', 'categorized_dataMay25', 'categorized_dataMay26', 'categorized_dataMay27', 'categorized_dataMay28']

# Initialize an empty list to store the category IDs
category_ids = []

# Read the data from each file and extract category IDs
for file_path in file_paths:
    with open(file_path, 'r') as file:
        data = json.load(file)
        category_ids.extend([item['snippet']['categoryId'] for item in data])

# Create a pandas DataFrame
df = pd.DataFrame(category_ids, columns=['Category ID'])

# Count the occurrences of each category ID
category_counts = df['Category ID'].value_counts()

# Select the top three categories
top_three_categories = category_counts.head(3)

# Generate a bar plot of the top three categories
top_three_categories.plot(kind='bar')

# Set plot title and axis labels
plt.title('Top Three Category IDs')
plt.xlabel('Category ID')
plt.ylabel('Count')

# Display the plot
plt.show()
```
