```python
import matplotlib.pyplot as plt
import pandas as pd
import requests
import json
import csv
import os

from urllib.parse import urlparse
```

```python
# This code will call Met Public API to get Metadate for the objects that are from Iran and then uses 
# PrimaryImageUrl to get the picture of the objects if it is available in Met database 


# URL to call to get object info from met Server
# This link provides more info: https://metmuseum.github.io/#search

# API call to get Irans object metadata
url = "https://collectionapi.metmuseum.org/public/collection/v1/search?geoLocation=Iran&q=*"
response = requests.get(url)
data = response.json()
object_ids = data["objectIDs"]


# Creating a path to store images in Destktop in MetImage folder 
folder_path = os.path.expanduser("~/Desktop/MetImage")  # Specify the folder path on your desktop

# Create the folder if it doesn't exist
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# For ech of the object if the Primary Image Url is available download the image 
for object_id in object_ids:
    url = f"https://collectionapi.metmuseum.org/public/collection/v1/objects/{object_id}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        primary_image_url = data.get("primaryImage", "")
        if primary_image_url:
            # Download the image
            image_response = requests.get(primary_image_url)
            if image_response.status_code == 200:
                filename = f"{object_id}.jpg"  # Set desired filename
                file_path = os.path.join(folder_path, filename)
                with open(file_path, "wb") as file:
                    file.write(image_response.content)
                print(f"Image downloaded for object ID {object_id}.")
            else:
                print(f"Failed to download image for object ID {object_id}.")
        else:
            print(f"No primary image found for object ID {object_id}.")
    else:
        print(f"Failed to fetch object with ID {object_id}.")
```
