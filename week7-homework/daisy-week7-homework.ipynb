```python
import pandas as pd

df = pd.read_csv('Code_Complaints_and_Violations.csv', low_memory=False)

# Convert the 'OpenDate' column to datetime format
df['OpenDate'] = pd.to_datetime(df['OpenDate'])

# Create a new time series using the 'OpenDate' column as the index and the 'OriginalZip' column as the values
ts = pd.Series(data=df['OriginalZip'].values, index=df['OpenDate'])

# Use resample() to get the number of complaints per week
ts_resampled = ts.resample('W').count()

# Plot the time series
ts_resampled.plot(figsize=(12,6))

print(ts_resampled)
```
