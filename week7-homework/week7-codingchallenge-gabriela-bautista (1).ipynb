```python
# Build a new dataset that includes only the "RecordType", "OriginalZip", and "Description" columns.

#record_zip_description_dataset = df[["RecordType","OriginalZip","Description"]]
record_zip_description_dataset = df.loc[:,["RecordType","OriginalZip","Description"]]
```

```python
# Use this second dataset to filter the dataset down to just rows from your zipcode. If you don't live in Seattle, you can just use my zip code (98112) which covers north Capitol Hill and Montlake or you can pick an area you think is interesting from this map.
cap_hill_dataset = record_zip_description_dataset[record_zip_description_dataset["OriginalZip"] == 98122]
print(cap_hill_dataset.head())
# Now look at the number and proportion of different types of records in this subset.

print(cap_hill_dataset["RecordType"].value_counts())
#df.groupby("RecordType")["RecordType"].count()
cap_hill_dataset["RecordType"].hist()
```

```python
# Be ready to explain if the distribution in this zipcode different than the distribution in Seattle overall? If not, how is it different?
# The same but with less numbers

# Once again, print out the "Description" of several examples from each category. What kinds of things are included?
for record_type in record_types:
    result=cap_hill_dataset[cap_hill_dataset['RecordType'] == record_type]["Description"].head()
    for r in result:
        print(record_type)
        print(r)
        print()
```

```python
# Use pandas to write out the three-column dataset to TSV (with tabs instead of commas).
file_name = "cap_hill_dataset.tsv"
cap_hill_dataset.to_csv(file_name, sep='\t', index=False)
print(f"{file_name} created")
```
