<b> It's about time

Create a new timeseries (use a pandas Series) that contains zip code and that uses the "OpenDate" column as the index. Be sure to check the type of the "OpenDate" column and make sure it's in the pandas datetime format.

```python
complaints_violations['OpenDate'] = pd.to_datetime(complaints_violations['OpenDate'])
complaints_violations.set_index('OpenDate', inplace=True)
complaints_violations_zipcode = complaints_violations['OriginalZip']
```

```python
complaints_violations_zipcode
```

Use the .resample() function associated with your pandas time series so that it shows the number of complaints per week overall and visualize this with a time series plot.

```python
weekly_complaints_violations_zipcode = complaints_violations_zipcode.resample('W').count()
```

```python
weekly_complaints_violations_zipcode
```

```python
weekly_complaints_violations_zipcode.plot()
```
