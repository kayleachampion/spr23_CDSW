I want to know how many claims was open per year

```python
import pandas as pd

# Read the dataset into a pandas dataframe
df = claims

# Convert the date column to datetime format
df['OpenDate'] = pd.to_datetime(df['OpenDate'])

# Group the data by year and count the number of events
events_per_year = df.groupby(df['OpenDate'].dt.year)['OpenDate'].count()

# Print the result
print(events_per_year)
```
