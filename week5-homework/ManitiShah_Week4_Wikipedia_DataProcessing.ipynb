### Identify a famous person who has been famous for at least a few years and that you have some personal interest in. Use the Wikimedia API to collect page view data from the English Wikipedia article on that person. Now use that data to generate a time-series visualization and include a link to it in your notebook.

```python
#processing the data collected

with open("Harry_Styles_Page_Views.json", 'r') as input_file:
    input_data = input_file.read()
    
EnglishData = json.loads(input_data)
```

```python
EnglishData['items'][3]
```

```python
def clean_up_timestamp(day):
    new_time_stamp = day[0:4] + "-" + day[4:6] + "-" + day[6:8]
    return new_time_stamp
```

```python
monthly_views = {} 

for month_dict in EnglishData['items'] :
    day = clean_up_timestamp(month_dict['timestamp'])
    views = month_dict['views']
    monthly_views[day] = month_dict['views']
```

```python
print(monthly_views)
```

```python
#writing the data to the file 

with open('view_by_month_hs.tsv', 'w') as output_file:
    print("month\tmonthlt views", file=output_file)
    for day in monthly_views:
        print(f"{day}\t{monthly_views[day]}", file=output_file)
```

### Sum up all of the page views of Alternative Rock Bands with a TSV file with these total numbers.

```python
band_dicts = []
with open("AlternativeRockBands.jsonl", 'r') as input_file:
    for line in input_file.readlines():
        new_data = json.loads(line)
        band_dicts = band_dicts + new_data['items']
```

```python
band_dicts
```

```python
TotalViews = {}

for monthlydata in band_dicts:
    bandname = monthlydata['article']
    date = clean_up_timestamp(monthlydata['timestamp'])
    monthlyviews = monthlydata['views']
    
    if bandname not in TotalViews:
        TotalViews[bandname] = monthlyviews
    else:
        TotalViews[bandname] += monthlyviews
        
print(TotalViews)
    
    
    
```

```python
with open ("ViewsARB.tsv", 'w') as my_file :
    print('bandname\ttotalviews', file=my_file)
    for band in TotalViews:
        print(f"{band}\t{TotalViews[band]}", file=my_file)
```
