```python
def parse_wikimedia_timestamp (old_date_string):
    new_date_string = f"{old_date_string[0:4]}/{old_date_string[4:6]}/{old_date_string[6:8]}"
    return new_date_string
```

```python
for item in strat_data['items']:
    date = item["timestamp"]
    print(parse_wikimedia_timestamp(date), "\t", item['views'])
```

```python
strat_items = strat_data['items']

tele_items = tele_data['items']

strat_dict = {}

tele_dict = {}

for item in strat_items:
    strat_dict[f"{item['timestamp']}"] = item['views']
    
for item in tele_items:
    tele_dict[f"{item['timestamp']}"] = item['views']
    
with open(r"C:\Users\Nicke\OneDrive\Desktop\CommunityDataScience\fender_strat&tele_views.tsv", "w") as fenderwiki_file:
    print("date\tstrat_views\ttele_views", file=fenderwiki_file)  
    
    for date in strat_dict.keys():   
        strat_views = strat_dict[date]
        tele_views = tele_dict[date]
        day_date = parse_wikimedia_timestamp(date)
    
        print(f"{day_date}\t{strat_views}\t{tele_views}", file=fenderwiki_file)
```

```python
import requests
import json
import time

headers = {'User-Agent': 'Mozilla/5.0' , 'From': 'neah@uw.edu', 'accept': 'application/json'}

with open(r"C:\Users\Nicke\OneDrive\Desktop\CommunityDataScience\washington_alternative_bands.tsv", "w") as bandswiki_file:
    print("page_name\tdate\tpage_views", file=bandswiki_file) 

    with open(r"C:\Users\Nicke\OneDrive\Desktop\CommunityDataScience\washington_alternative_bands.jsonl", 'r') as f:

        for line in f.readlines():
            line_dict = json.loads(line)
        
            page_title = line_dict['page_title']
        
            #print(page_title)

            url = f"https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia.org/all-access/user/{page_title}/monthly/20150101/20230429"

            band_response = requests.get(url, headers = headers)

            band_json = json.loads(band_response.text)

            band_dict = band_json['items']

            for views_data in band_dict:
                print(f"{page_title}\t{parse_wikimedia_timestamp(views_data['timestamp'])}\t{views_data['views']}", file=bandswiki_file)
```
