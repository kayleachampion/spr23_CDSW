```python
import requests
import json
import csv
from datetime import datetime

# Set the API endpoints for the English, German, and French language editions
endpoints = {
    'en': 'https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia/all-access/all-agents/Tiger_Woods/daily/2015010100/2023043000',
    'de': 'https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/de.wikipedia/all-access/all-agents/Tiger_Woods/daily/2015010100/2023043000',
    'fr': 'https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/fr.wikipedia/all-access/all-agents/Tiger_Woods/daily/2015010100/2023043000'
}

# Set the headers for the API request
headers = {
    'User-Agent': 'Chrome/89.0.4389.82 Safari/537.36',
    'From': 'herreraleopoldo12@gmail.com',
    'Api-User-Agent': 'Example/1.0'
}

# Make the API requests and parse the JSON response data
data = {}
for lang, endpoint in endpoints.items():
    response = requests.get(endpoint, headers=headers)
    items = json.loads(response.text)['items']
    for item in items:
        timestamp = datetime.strptime(item['timestamp'], '%Y%m%d%H')
        date = timestamp.date()
        views = item['views']
        if date not in data:
            data[date] = {lang: views}
        else:
            data[date][lang] = views

# Write the page views data to a CSV file
with open('tiger_woods_page_views_3Languages.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Date', 'English', 'German', 'French'])
    for date, views in data.items():
        writer.writerow([date, views.get('en', 0), views.get('de', 0), views.get('fr', 0)])
        
# https://docs.google.com/spreadsheets/d/1qaG0nlYwDLFJ7Z7Mq3Q2nwiED-qh7FXiVUCcXjqF1cU/edit?usp=sharing
```

```python
#3.1
import requests
import json
from datetime import datetime

# Set the API endpoints for the English language editions of the Marvel Comics and DC Comics pages
endpoints = {
    'marvel': 'https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia/all-access/all-agents/Marvel_Comics/daily/2022010100/2022013100',
    'dc': 'https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia/all-access/all-agents/DC_Comics/daily/2022010100/2022013100'
}

# Set the headers for the API request
headers = {
    'User-Agent': 'Chrome/89.0.4389.82 Safari/537.36',
    'From': 'herreraleopoldo12@gmail.com',
    'Api-User-Agent': 'Example/1.0'
}

# Make the API requests and parse the JSON response data
data = {}
for publisher, endpoint in endpoints.items():
    response = requests.get(endpoint, headers=headers)
    items = json.loads(response.text)['items']
    for item in items:
        timestamp = datetime.strptime(item['timestamp'], '%Y%m%d%H')
        date = timestamp.date()
        views = item['views']
        if date not in data:
            data[date] = {publisher: views}
        else:
            data[date][publisher] = views

# Calculate the total views for each publisher in 2022
marvel_views = 0
dc_views = 0
for date, views in data.items():
    if date.year == 2022:
        marvel_views += views.get('marvel', 0)
        dc_views += views.get('dc', 0)

# Print the results
if marvel_views > dc_views:
    print('Marvel Comics received more views on English Wikipedia in 2022')
elif dc_views > marvel_views:
    print('DC Comics received more views on English Wikipedia in 2022')
else:
    print('Marvel Comics and DC Comics received the same number of views on English Wikipedia in 2022')
```

```python
# 3.2
import requests
import json
import csv
from datetime import datetime

# Set the API endpoints for the Marvel and DC Comics pages in English Wikipedia
endpoints = {
    'Marvel Comics': 'https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia/all-access/all-agents/Marvel_Comics/daily/2015010100/2023043000',
    'DC Comics': 'https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia/all-access/all-agents/DC_Comics/daily/2015010100/2023043000'
}

# Set the headers for the API request
headers = {
    'User-Agent': 'Chrome/89.0.4389.82 Safari/537.36',
    'From': 'herreraleopoldo12@gmail.com',
    'Api-User-Agent': 'Example/1.0'
}

# Make the API requests and parse the JSON response data
data = {}
for page, endpoint in endpoints.items():
    response = requests.get(endpoint, headers=headers)
    items = json.loads(response.text)['items']
    for item in items:
        timestamp = datetime.strptime(item['timestamp'], '%Y%m%d%H')
        date = timestamp.date()
        views = item['views']
        if date not in data:
            data[date] = {page: views}
        else:
            data[date][page] = views

# Write the page views data to a CSV file
with open('marvel_dc_page_views.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Date', 'Marvel Comics', 'DC Comics'])
    for date, views in data.items():
        writer.writerow([date, views.get('Marvel Comics', 0), views.get('DC Comics', 0)])
        
#https://docs.google.com/spreadsheets/d/1xuXA3f4AiXO-meGmygu0M8_JTBoT_5THQKiW866zO5c/edit?usp=sharing
```

```python
import requests
import json
from datetime import datetime

def get_data_from_api(url):
    response = requests.get(url)
    data = json.loads(response.text)['items']
    pageviews = {}
    for item in data:
        date = datetime.strptime(item['timestamp'][:6], '%Y%m')
        if date.year >= 2015 and date.year <= 2023:
            if date.year not in pageviews:
                pageviews[date.year] = {}
            pageviews[date.year][date.month] = item['views']
    return pageviews

marvel_url = "https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia.org/all-access/user/Marvel_Comics/monthly/2015010100/2023123100"
dc_url = "https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia.org/all-access/user/DC_Comics/monthly/2015010100/2023123100"

marvel_pageviews = get_data_from_api(marvel_url)
dc_pageviews = get_data_from_api(dc_url)

years_when_less_popular_page_received_more_views = []
for year in range(2015, 2024):
    marvel_views = sum(marvel_pageviews[year].values())
    dc_views = sum(dc_pageviews[year].values())
    if marvel_views < dc_views:
        years_when_less_popular_page_received_more_views.append(year)

print(years_when_less_popular_page_received_more_views)
```
