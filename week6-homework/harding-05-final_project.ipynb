```python
my_response = requests.get("https://api.army.mil/api/v1/social/twitter/posts?user=%40usarmy&limit=10000")
if my_response.status_code == 200:
    data = my_response.json()
    
readiness = 0

for tweet in data:
    print(tweet['caption'])
    print()
    if 'readiness' in tweet['caption']:
        readiness += 1
        
#print(readiness)
        
#print(data)

with open("usarmy_tweets.json", 'w') as my_file:
    data_string = json.dumps(data)
    print(data_string, file=my_file)
```
