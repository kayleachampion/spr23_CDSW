```python
import json
```

### Cleaning Twilight WikiData 

```python
with open ('WikiData.jsonl', 'r') as input_file :
    master = input_file.readlines()
```

```python
WikiData = {}
anoncount = 0
revisioncount = 0 
minorcount = 0

for datachunk in master:
    data = json.loads(datachunk)

    #Let's get the list of pages 
    pages = data['query']['pages']

    #Let's get the name of those pages 
    for page_id in pages:
        title = pages[page_id]['title']
        revisions = pages[page_id]['revisions']
       
        for singlerevision in revisions:
            revision = singlerevision['revid']
            revisioncount += 1
            user = singlerevision['user']
            size = singlerevision['size']
            time = singlerevision['timestamp']
            
            if title not in WikiData.keys():
                WikiData[title] = [singlerevision]
            else:
                WikiData[title].append(singlerevision)
            
            
            if 'anon' in singlerevision.keys():
                anoncount += 1
                
            if 'minor' in singlerevision.keys():
                minorcount += 1
                

print(revisioncount)
print(anoncount)
print(minorcount)           
```

```python
def clean_up_timestamp(timestamp): 
    return (timestamp.split("T")[0])
```

```python
with open ('CleanWikiData.tsv', 'w') as output_file :
    print ("title\tRevision Id\tsize\ttimestamp\tuser", file=output_file)
    for title in WikiData.keys():
           for revisions in WikiData[title]:
                if 'anon' in revisions.keys():
                    user = 'ANON'
                else:
                    user = 'NOT ANON'
                print(f"{title}\t{revisions['revid']}\t{revisions['size']}\t{clean_up_timestamp(revisions['timestamp'])}\t{user}", file=output_file)
```

### What proportion of those edits were made by users without accounts ("anon") (Wikipedia)

```python
#We already have Data on total revisions and anon revisions 

prop_anon = anoncount / revisioncount 

print(f"total edits: {revisioncount}")
print(f"anon edits: {anoncount}")
print(f"proportion anon: {prop_anon}")
```

### What proportion of those edits were marked as ("minor") (Wikipedia)

```python
#We already have Data on total revisions and minor revisions 

prop_minor = minorcount / revisioncount 

print(f"total edits: {revisioncount}")
print(f"minor edits: {minorcount}")
print(f"proportion anon: {prop_minor}")
```
