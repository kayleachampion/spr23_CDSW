```python
girl_initials = {}

for name in girls.keys():
    initial = name[0]
    if initial in girl_initials.keys():
        girl_initials[initial] += 1
    else:
        girl_initials[initial] = 1
        
print(girl_initials)
```

```python
with open("girl_initials.tsv", "w") as my_file:
    print("initial\tnumber", file = my_file)
    for initial in girl_initials.keys():
        print(f"{initial}\t{girl_initials[initial]}", file = my_file)
```

#How many girl names are not used for boy names? And what is the precentage?

```python
girl_names = list(girls.keys())
boy_names = list(boys.keys())
girl_names_not_for_boys = []

for name in girl_names:
    if name not in boy_names:
        girl_names_not_for_boys.append(name)

num_girl_names_not_for_boys = len(girl_names_not_for_boys)
precentage = num_girl_names_not_for_boys / len(girls.keys()) * 100
print (str(precentage) + " % of girl names are not used as boy names in 2021")
```
