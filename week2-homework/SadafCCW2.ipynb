```python
#Find and print all the words that have all 5 vowels in alphabetical order.

for word in wordlist:
    counter = 0
    
    for letter in word:
        if letter in "aeiou":
            counter = counter + 1
        
    print (f"{word} has {counter} vowels!")
```

```python
#Find and print the longest word that has no vowels

longestWord =  max(wordlist, key=len)
print (longestWord)
len(longestWord)
```

```python
#Find and print the words that contain 4 or more 'l's.
#How many are there?

lList = []

for word in wordlist:
    lCounter = 0
    for letter in word:
        if letter == "l":
            lCounter = lCounter + 1
    if lCounter >= 4:
        print(f"{word} has {lCounter} Ls!")
        lList.append (word)
        

```

```python
#Print out a list of every palindrome (a word that is the same reversed) that's over 6 letters in length.
#How many are there?

palindromes = []

for word in wordlist:
    if len(word) > 6:
        if word == word[::-1]:
            palindromes.append(word)
    
len(palindromes)
    
```

```python
#10 letter words spelt on the top line of a QWERTY keyboard
choice_letters = ["q", 'w', 'e', "r", "t", "y"]

matches = []
for word in wordlist:
    has_all_letters = True
    for choice_letter in choice_letters:
        if choice_letter not in word:
            has_all_letters = False
    if has_all_letters:
        matches.append(word)   
```
