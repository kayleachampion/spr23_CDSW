# 1

```python
# pt 1: end in "well"

count = 0
for word in wordlist:
    if word[-4:] == "well":
        print(word)
        count = count + 1
print(count, "words end in well!")

print()

# pt 2:

count = 0
for word in wordlist:
    if word[-2:] == "mt":
        print(word)
        count = count + 1
print(count, "words end in mt!")

print()

# pt 3:

count = 0
for word in wordlist:
    if word[-3:] == "gry":
        print(word)
        count = count + 1
print(count, "words end in gry!")
```

# 2

```python
# pt 1

vowels = ['a', 'e', 'i', 'o', 'u']

for vowel in vowels:
    count = 0
    for word in wordlist:
        if word[0] == vowel:
            count = count + 1
    print(count, 'words start with', vowel)

print("most common start is 'a'!")    

print()

# pt 2

for vowel in vowels:
    count = 0
    for word in wordlist:
        if word[-1] == vowel:
            count = count + 1
    print(count, 'words end with', vowel)

print("most common end is 'e'!")
```

```python
# pt 3

count = 0
for word in wordlist:
    a_pos = word.find('a')
    if a_pos >= 0:
        e_pos = word.find('e', a_pos + 1)
        if e_pos >= 0:
            i_pos = word.find('i', e_pos + 1)
            if i_pos >= 0:
                o_pos = word.find('o', i_pos + 1)
                if o_pos >= 0:
                    u_pos = word.find('u', o_pos + 1)
                    if u_pos >= 0:
                        print(word)
                        count = count + 1
print("found", count)

print()

# this is very similar, but uses 'continue' to avoid a big tree

count = 0
for word in wordlist:
    a_pos = word.find('a')
    if a_pos == -1:
        continue
    e_pos = word.find('e', a_pos + 1)
    if e_pos == -1:
        continue
    i_pos = word.find('i', e_pos + 1)
    if i_pos == -1:
        continue
    o_pos = word.find('o', i_pos + 1)
    if o_pos == -1:
        continue
    u_pos = word.find('u', o_pos + 1)
    if u_pos == -1:
        continue
    print(word)
    count = count + 1
print("found", count)

print()

# alternative approach, using a scan

count = 0
for word in wordlist:
    next_vowel = 'a'
    for char in word:
        if char == next_vowel:
            if next_vowel == 'a':
                next_vowel = 'e'
            elif next_vowel == 'e':
                next_vowel = 'i'
            elif next_vowel == 'i':
                next_vowel = 'o'
            elif next_vowel == 'o':
                next_vowel = 'u'
            elif next_vowel == 'u':
                print(word)
                count = count + 1
                next_vowel = '*'
print("found", count)

# last, here's the wizardly way

import re

count = 0
for word in wordlist:
    if re.match('.*a.*e.*i.*o.*u.*', word):
        print(word)
        count = count + 1
print("found", count)
```

```python
longest_word_with_no_vowels_so_far = ""

for word in wordlist:
    
    if (not ('a' in word) and not ('e' in word) and not ('i' in word) and
        not ('o' in word) and not ('u' in word)):
        
        if len(word) > len(longest_word_with_no_vowels_so_far):
            longest_word_with_no_vowels_so_far = word

print(longest_word_with_no_vowels_so_far)
```

# 4

```python
# pts 1  & 2

count_of_words = 0
for word in wordlist:
    count_of_ls = 0
    for char in word:
        if char == 'l':
            count_of_ls = count_of_ls + 1
    if count_of_ls >= 4:
        print(word)
        count_of_words = count_of_words + 1
print("found", count_of_words, "words with 4 or more 'l's!")
```

# 6

```python
count = 0
for word in wordlist:
    if len(word) > 6 and word[::-1] == word:
        print(word)
        count = count + 1
print(count, "palindromes found!")
```

```python
# let's do "10 letter words spelt on the top line of a QWERTY keyboard"

top_line = ['q', 'w', 'e', 'r', 't', 'y', 'u', 'i', 'o', 'p']
for word in wordlist:
    if len(word) == 10:
        count_bad_letters = 0
        for char in word:
            if not char in top_line:
                count_bad_letters = count_bad_letters + 1
        if count_bad_letters == 0:
            print(word)
```
