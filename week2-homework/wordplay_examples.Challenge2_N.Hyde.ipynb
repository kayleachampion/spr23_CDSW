### Example 1: Let's come up with a program to tell us how many words start with "aa"

```python
matches = []
for word in wordlist:
    if word[0:2] == "aa":
        matches.append(word)
len(matches)
```

```python
for match in matches:
    print(match)
```

## Example 2: print every word that contains two "l"s in a row 

```python
matches = []

substring = "ll"
# Print all words containing 'l'.
for word in wordlist:
    if substring in word:
        matches.append(word)
len(matches)
```

```python
print(matches)
```

## Example 3: print out a word if it contains an x, y, or z

```python
choice_letters = ['x', 'y', 'z']

matches = []
for word in wordlist:
    has_all_letters = True
    for choice_letter in choice_letters:
        if choice_letter not in word:
            has_all_letters = False
    if has_all_letters:
        matches.append(word)    
```

```python
len(matches)
```

```python
print(matches)
```

2.1 Counting Vowels

```python
search_vowel = "a"

vowel_count = 0

for word in wordlist:
    if word[0] == search_vowel:
        
        vowel_count = vowel_count + 1
        
print (vowel_count, "words that begin with the vowel a")
```

```python
search_vowels = ["a", "e", "i", "o", "u"]

for vowel in search_vowels:
    vowel_count = 0

    for word in wordlist:
        if word[0] == vowel:
        
            vowel_count = vowel_count + 1
        
    print (vowel_count, "words that begin with the vowel ", vowel)
```

Alternative ~   

```python
search_vowels = ["a", "e", "i", "o", "u"]

for vowel in search_vowels:
    vowel_count = 0

    for word in wordlist:
        if word.startswith(vowel):      #returning a boolean with startswith
        
            vowel_count = vowel_count + 1
        
    print (vowel_count, "words that begin with the vowel ", vowel)
```
