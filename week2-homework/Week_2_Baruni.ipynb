### Example 1: Let's come up with a program to tell us how many words start with "aa"

```python
matches = []
for word in wordlist:
    if word[0:2] == "aa":
        matches.append(word)
len(matches)
```

```python
for match in matches:
    print(match)
```

## Example 2: print every word that contains two "l"s in a row 

```python
matches = []

substring = "ll"
# Print all words containing 'l'.
for word in wordlist:
    if substring in word:
        matches.append(word)
len(matches)
```

```python
print(matches)
```

## Example 3: print out a word if it contains an x, y, and z

```python
choice_letters = ['x', 'y', 'z']

matches = []
for word in wordlist:
    has_all_letters = True
    for choice_letter in choice_letters:
        if choice_letter not in word:
            has_all_letters = False
    if has_all_letters:
        matches.append(word)    
```

```python
len(matches)
```

```python
print(matches)
```

```python
# Solution 2

#How many words end with each Vowel

# e is the most common vowel that ends a word

vowels = ['a', 'e', 'i', 'o', 'u']

for vowel in vowels:
    wordcount = 0
    for word in wordlist:
        if word[-1:] == vowel:
            wordcount = wordcount + 1
    print(str(wordcount) + " words end with the letter " + vowel)
```

```python
# Solution 2

#How many words have all 5 vowels in alphabetical order. Print all the words
# aaeiou
# aaeeeiou
hasallvowels = []
vowels = ['a', 'e', 'i', 'o', 'u']
previousVowel = 'a'

for word in wordlist:
    i = 0
    has_all_vowels= False
    for character in word:
        if character not in vowels:
            continue
        else:
            if character == vowels[i]:
                previousVowel = character
                i = i + 1
                if i == 5:
                    i = 4
                    has_all_vowels = True
            elif character == previousVowel:
                continue
            else:
                has_all_vowels = False
                break
                
    if has_all_vowels:
        hasallvowels.append(word)
print(hasallvowels)
print(len(hasallvowels))
```

```python
# Solution 3 - longest word with no vowels

longest_word = []
vowels = ['a', 'e', 'i', 'o', 'u']

for word in wordlist:
    has_no_vowels = True
    for vowel in vowels:
        if vowel in word:
            has_no_vowels = False
            break
    
    if has_no_vowels:
        longest_word.append(word)

print(longest_word)

#print the longest word that has no vowels
```

```python
# Solution 4 - words with 4 or more l's, how many are there?

l_words = []

for word in wordlist:
    count = 0
    for letter in word:
        if 'l' == letter:
            count = count + 1
        if count >= 4:
            l_words.append(word)

print(l_words)
print(len(l_words))
```

```python
#Solution 6 Print out a list of every palindrome (a word that is the same reversed) that's over 6 letters in length.
#How many are there?

palindromes = []

for word in wordlist:
    i = 0
    isPalindrome = True
    if len(word) >= 6:
        for i in range(0, len(word)//2):
            if word[i] != word[len(word) - i - 1]:
                isPalindrome = False
                break
            i = i + 1;
        if isPalindrome:
            palindromes.append(word)
print(palindromes)
print(len(palindromes))
```

```python
#Solution 6 Print out a list of every palindrome (a word that is the same reversed) that's over 6 letters in length.
#How many are there?

palindromes = []

for word in wordlist:
    i = 0
    isPalindrome = True
    if len(word) >= 6:
        if word == word[::-1]:
            palindromes.append(word)
print(palindromes)
print(len(palindromes))
```
