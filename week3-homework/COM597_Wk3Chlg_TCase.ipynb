```python
most_common_boys_initials = {}

for boy in boys:
    first_letter = boy[0]
    if first_letter in most_common_boys_initials.keys():
        most_common_boys_initials[first_letter] += 1
    else:
        most_common_boys_initials[first_letter] = 1
        
print(most_common_boys_initials)
```

```python
boy_initials = {}

for boy in boys:
    first_letter = boy[0]
    if first_letter in boy_initials.keys():
        boy_initials[first_letter] += 1
        
    else:
        boy_initials[first_letter] = 1

for first_letter in sorted(boy_initials.keys()):
    print(first_letter, ":", boy_initials[first_letter])
```

```python
most_common_girls_initials = {}

for girl in girls:
    first_letter = girl[0]
    if first_letter in most_common_girls_initials.keys():
        most_common_girls_initials[first_letter] += 1
        
    else:
        most_common_girls_initials[first_letter] = 1
    
print(most_common_girls_initials)
```

```python
girl_initials = {}

for girl in girls:
    first_letter = girl[0]
    if first_letter in girl_initials.keys():
        girl_initials[first_letter] += 1
        
    else:
        girl_initials[first_letter] = 1

for first_letter in sorted(girl_initials.keys()):
    print(first_letter, ":", girl_initials[first_letter])
```

```python
with open("girl_initials.tsv", "w") as my_file:
    print("letter\tcount", file = my_file)
    for initial in girl_initials.keys():
        print(f"{initial}\t{girl_initials[initial]}", file = my_file)
```

```python
with open("boy_initials.tsv", "w") as my_file:
    print("letter\tcount", file = my_file)
    for initial in boy_initials.keys():
        print(f"{initial}\t{boy_initials[initial]}", file = my_file)
```

```python
for initial in girl_initials:
    if girl_initials[initial] > boy_initials[initial]:
        print(initial + ": " + "there are more girl initials")
    else:
        print(initial + ": " + "there are more boy initials")
```
