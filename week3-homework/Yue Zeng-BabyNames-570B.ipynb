# 4 She wasn't long for this dataset

```python
longest = 0
longest_name = []
for name in whole.keys():
    if len(name) > longest:
        longest = len(name)
        longest_name.clear()
        longest_name.append(name)
    elif len(name) == longest:
        longest_name.append(name)
print(f"the longest name in the dataset are {', '.join(longest_name)}")
total_boys = 0
for name in longest_name:
    total_boys += boys.get(name, 0)
print(f"There is {total_boys} boys names are exactly that length")

total_girls = 0
for name in longest_name:
    total_girls += girls.get(name, 0)
print(f"There is {total_girls} girls names are exactly that length")
```

# 5 Initials to spreadsheets

```python
girl_initials = {c: 0 for c in "abcdefghijklmnopqrstuvwxyz"}
boys_initials = {c: 0 for c in "abcdefghijklmnopqrstuvwxyz"}
for name in girls.keys():
    girl_initials[name[0]] += 1

for name in boys.keys():
    boys_initials[name[0]] += 1

with open("girl_initials.tsv", "w") as out:
    out.write(f"letter\tnumber\n")
    for c in "abcdefghijklmnopqrstuvwxyz":
        out.write(f"{c}\t{girl_initials[c]}\n")

with open("boys_initials.tsv", "w") as out:
    out.write(f"letter\tnumber\n")
    for c in "abcdefghijklmnopqrstuvwxyz":
        out.write(f"{c}\t{boys_initials[c]}\n")

letter = "s"
if girl_initials[letter] < boys_initials[letter]:
    print(f"there are more boys names for letter '{letter}'")
elif girl_initials[letter] > boys_initials[letter]:
    print(f"there are more girls names for letter '{letter}'")
else:
    print(f"there are same names for letter '{letter}'")
```

# 6 Concentration in names

```python
most_popular_boys_names = sorted(boys.keys(), key=lambda name: boys[name], reverse=True)[:10]
total_boys = sum(boys.values())
boys_have_most_popular_boys_names = sum([boys[name] for name in most_popular_boys_names])
print(f"the percentage of boys have one of the 10 most popular boys names is {boys_have_most_popular_boys_names / total_boys * 100}%")

most_popular_girls_names = sorted(girls.keys(), key=lambda name: girls[name], reverse=True)[:10]
total_girls = sum(girls.values())
boys_have_most_popular_girls_names = sum([girls[name] for name in most_popular_girls_names])
print(f"the percentage of girls have one of the 10 most popular boys names is {boys_have_most_popular_girls_names / total_girls * 100}%")


top_10_percent_most_popular_boys_names = sorted(boys.keys(), key=lambda name: boys[name], reverse=True)[:int(len(boys) * 0.1)]
girls_count = sum([girls.get(name, 0) for name in top_10_percent_most_popular_boys_names])
print(f"there are {girls_count} girls were given one of those names")

top_10_percent_most_popular_girls_names = sorted(girls.keys(), key=lambda name: girls[name], reverse=True)[:int(len(girls) * 0.1)]
boys_count = sum([boys.get(name, 0) for name in top_10_percent_most_popular_girls_names])
print(f"there are {boys_count} boys were given one of those names")
```
