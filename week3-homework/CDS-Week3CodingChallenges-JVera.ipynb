```python
#3 A sense of what's common

#What is the most common name for each gender in your data (i.e. 2021)?

for name in boys.keys():
    if boys[name] == max(boys.values()):
        print ("the most common name for boys is " + name)
    
for name in girls.keys():
    if girls[name] == max(girls.values()):
        print ("the most common name for girls is " + name)

#What is the least common name?

for name in boys.keys():
    if boys[name] == min(boys.values()):
        print ("the least common name for boys is " + name)
         
for name in girls.keys():
    if girls[name] == min(girls.values()):
        print ("the least common name for girls is " + name)
        

#How often do the least common names occur? (Does your answer to this question bother you? Why?)
# 'Often' refers to frequency. Are we looking for an incident rate for 2021?

# For boys
rate_boys = min(boys.values()) / len(boys)
print(rate_boys)

# For boys
rate_girls = min(girls.values()) / len(girls)
print(rate_girls)
```

```python
#4 She wasn't long for this dataset

# What is the longest name in the dataset? 

#First, I need to combine the dictionaries
babynames_dict = boys.copy()
babynames_dict.update(girls)
#print(babynames_dict)

#longestname, total dictionary
print(max(babynames_dict, key=len) + " is the longest name in the whole dictionary")

#longest name in boys
print(max(boys.keys(), key=len) + " is the longest boys name") #ok that checks out

#longest name in girls
print(max(girls.keys(), key=len) + " is the longest girls name") #ok but there may be multiple with that many letters?
```

```python
#5 Initials to Spreadsheets

#Make a dictionary girl_initials that says, for each letter of the alphabet (a-z), 
#how many unique girl names begin with that letter.

girl_initials = {}

for name in girls.keys():
    if name[0] in girl_initials.keys():
        girl_initials[name[0]] = girl_initials[name[0]] + 1
    else:
        girl_initials[name[0]] = 1
    
print(girl_initials)

#Explanation: for each name in the list of keys in the list of girl names, if the 0 position letter in the name
#in girls_initials.keys, which is the empty list at this point... if that zero position is true, take that and add
#one. else, just set the zero position of the name of girl_intials to 1. Iterate through the loop.

#Do the same for boy names to make boy_initials.

boy_initials = {}

for name in boys.keys():
    if name[0] in boy_initials.keys():
        boy_initials[name[0]] = boy_initials[name[0]] + 1
    else:
        boy_initials[name[0]] = 1
    
print(boy_initials)


#Create a "tab separated values" (TSV) file that reports the data in girl_initials. 
#Be sure to include a descriptive header columns! You will probably want the file to end with
#.tsv so that your computer knows it's a TSV file.
import csv

tsvFileName = 'JV_WE3_girl_initials.tsv'

tsv_header_name = ['name', 'num of names']

each_row = girl_initials

with open(tsvFileName, 'w') as tsvfile:
    #csv writer to write in tsv file
    tsv_writer = csv.writer(tsvfile, delimiter='\t')
    #write header in tsv file
    tsv_writer.writerow(tsv_header_name)
    #write rows
    for key in each_row.keys():
        tsv_writer.writerow([key, each_row[key]])
    
    #close csv file
    tsvfile.close()


#Now do the same for boy_initials (be sure to save it into a different file!)

tsvFileName = 'JV_WE3_boy_initials.tsv'

tsv_header_name = ['name', 'num of names']

each_row = boy_initials

with open(tsvFileName, 'w') as tsvfile:
    #csv writer to write in tsv file
    tsv_writer = csv.writer(tsvfile, delimiter='\t')
    #write header in tsv file
    tsv_writer.writerow(tsv_header_name)
    #write rows
    for key in each_row.keys():
        tsv_writer.writerow([key, each_row[key]])
    
    #close csv file
    tsvfile.close()


#Once you've done this, load up the two files into Google Sheets or Excel.
#For every letter, be ready to tell if there are more boys names or girls names.

### Sorted lists. More girl names start with the letter A. 

#Play around with graphing and see if you can build some instructive graph that shows us something.
#Note: Obviously, you won't be able to include your Google Sheet result into your notebook. 
#Instead, please put it online somewhere (e.g., in Google Drive, or OneDrive, or Dropbox or similar) 
#create a link for sharing that doesn't requiring signing in, and put that link into your notebook so 
#we can click through and look at it!

###GIRL NAMES URL: https://docs.google.com/spreadsheets/d/1O4iEUaGw-LINqvlppjq-smM2NQqCRIrwba2OCSqL4Co/edit?usp=sharing

###BOY NAMES URL: https://docs.google.com/spreadsheets/d/14TJYPRuuR9NyviOxLiVcQ-408_u7AnhxvaxjL4K_cO4/edit?usp=sharing
```

```python
#6 Concentration in names
#What percentage of boys have one of the 10 most popular boys names? 

###

#Top 10 most popular boys names
top10boys = []
for value in sorted(boys.values(), reverse=True)[:10]:
    top10boys.append(value)
#print(top10boys)

#count up the number of top 10 names' values
numtopten = sum(top10boys)
#print(numtopten)

#count up the total number of values
numallboys = sum(boys.values())
#print(numallboys)

#do some math
maths = numtopten/numallboys
#print(maths)

#make this a percentage
percentage = "{:.0%}".format(maths)

print ("about " + percentage + " of boys have one of the 10 most popular boys names")


#What percentage of girls have one of the 10 most popular girls names?

#Top 10 most popular girls names
top10girls = []
for value in sorted(girls.values(), reverse=True)[:10]:
    top10girls.append(value)
#print(top10girls)

#count up the number of top 10 names' values
numtoptengirls = sum(top10girls)
#print(numtoptengirls)

#count up the total number of values
numallgirls = sum(girls.values())
#print(numallgirls)

#do some math
mathsgirls = numtoptengirls/numallgirls
#print(mathsgirls)

#make this a percentage
percentagegirls = "{:.0%}".format(mathsgirls)

print ("about " + percentagegirls + " of girls have one of the 10 most popular girls names")


#Take the top 10% most popular boys names. (For instance: If there were 500 boys names, we're 
#looking for the 50 most popular ones.) How many girls were given one of those names? 

#What is the top 10%?
totalboysnames = len(boys.keys())
#print(totalboysnames) #gives the length of the list, 13,993

#maths
tenpercent = 0.01
toptenpercentallboysnames = totalboysnames * tenpercent
#print(toptenpercentallboysnames) #139.93

#So, we're looking for the 140 (139.93 rounded up) most popular Values....
most_popular_boys = sorted(boys, key=boys.get, reverse=True)[:140]
#print(most_popular_boys)

#Take the top 10% most popular girls names. How many boys were given one of those names?

#What is the top 10%?
totalgirlsnames = len(girls.keys())
#print(totalgirlsnames) #gives the length of the list, 17,544

#maths
tenpercent = 0.01
toptenpercentallgirlsnames = totalgirlsnames * tenpercent
#print(toptenpercentallgirlsnames) #175.44

#So, we're looking for the 175 (175.44 rounded up) most popular Values....
most_popular_girls = sorted(girls, key=girls.get, reverse=True)[:175]
#print(most_popular_girls)

#Find the commonalities
commonalities = []

for name in most_popular_girls:
        count = 0
        if name in most_popular_boys:
            commonalities.append(name)
        else:
            count = count +1
print("The names that share lists are " + str(commonalities))
```
