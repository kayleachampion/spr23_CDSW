7.0 Total amount of matching and unmatching names.

```python
total_matching_names = 0

for name in boys.keys():
    if name in girls.keys():
        total_matching_names = total_matching_names + 1
        
print("There are", total_matching_names, "matching names in the dataset.")
```

```python
total_unmatched_names = 0

for name in boys.keys():
    if name not in girls.keys():
        total_unmatched_names += 1

print("There are", total_unmatched_names, "names in the dataset that are unique to boys and girls.")
```
