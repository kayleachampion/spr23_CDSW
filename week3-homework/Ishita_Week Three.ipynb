```python
#sourced from chatgpt
from ssadata import boys, girls

top_boys_names = list(boys.keys())[:10]
top_girls_names = list(girls.keys())[:10]

total_boys = sum(boys.values())
total_top_boys = sum(boys[name] for name in top_boys_names)
percentage_top_boys = 100 * total_top_boys / total_boys
print(f"Percentage of boys with a top name: {percentage_top_boys:.2f}%")

total_girls = sum(girls.values())
total_top_girls = sum(girls[name] for name in top_girls_names)
percentage_top_girls = 100 * total_top_girls / total_girls
print(f"Percentage of girls with a top name: {percentage_top_girls:.2f}%")
```
