```python
#Solution 5

girls_initials = {}


for name in girls.keys():
    first_letter = name[0]
    if first_letter in girls_initials.keys():
        girls_initials[first_letter] += 1
        
    else:
       girls_initials[first_letter] = 1
    
print(girls_initials)
```

```python
boys_initials = {}


for name in boys.keys():
    first_letter = name[0]
    if first_letter in boys_initials.keys():
        boys_initials[first_letter] += 1
        
    else:
       boys_initials[first_letter] = 1
    
print(boys_initials)
```
