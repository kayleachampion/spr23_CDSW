```python
#4 She wasn't long for this dataset
```

```python
#What is the longest name in the dataset? How many boys/girls names are exactly that length? What's going on?
```

```python
name_lengths_boys = {name: len(name) for name in boys}
longest_name_length_boys = max(name_lengths_boys.values())
num_boys_with_longest_name = list(name_lengths_boys.values()).count(longest_name_length_boys)

print('The longest name in the dataset has length:', longest_name_length_boys)
print('The number of boys with names of the longest length is:', num_boys_with_longest_name)
```

```python
longest_name_boys = []

for name in name_lengths_boys.keys():
    if name_lengths_boys[name] == 15:
        longest_name_boys.append(name)
print (longest_name_boys)
```

```python
name_lengths_girls = {name: len(name) for name in girls}
longest_name_length_girls = max(name_lengths_girls.values())
num_girls_with_longest_name = list(name_lengths_girls.values()).count(longest_name_length_girls)

print('The longest name in the dataset has length:', longest_name_length_girls)
print('The number of boys with names of the longest length is:', num_girls_with_longest_name)
```

```python
longest_name_girls = []

for name in name_lengths_girls.keys():
    if name_lengths_girls[name] == 14:
        longest_name_girls.append(name)
print (longest_name_girls)
```

```python
#5 Initials to spreadsheets
```

```python
girl_initials = {}

for name in girls.keys():
    for letter in name:
        if name[0] in girl_initials.keys():
            girl_initials[name[0]] += 1
        else:
            girl_initials[name[0]] = 1
print (girl_initials)
```

```python
boy_initials = {}

for name in boys.keys():
    for letter in name:
        if name[0] in boy_initials.keys():
            boy_initials[name[0]] += 1
        else:
            boy_initials[name[0]] = 1
print (boy_initials)
```

```python
with open("girls_initial.tsv", "w") as my_file:
    print("initial\tcount", file=my_file)
    for letter in girl_initials.keys():
        print(f"{letter}\t{girl_initials[letter]}", file=my_file)
```

```python
with open("boys_initial.tsv", "w") as my_file:
    print("initial\tcount", file=my_file)
    for letter in boy_initials.keys():
        print(f"{letter}\t{boy_initials[letter]}", file=my_file)
```
