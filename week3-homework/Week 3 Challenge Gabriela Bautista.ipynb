#2 Every baby counts
How many boy names and girl names are described in the dataset?
How many boys and girls (actual babies!) are described in the dataset?

```python
countboys = 0
countgirls = 0
boy_names = len(boys.keys())
girl_names = len(girls.keys())

for babyboy in boys.keys():
    countboys = countboys + boys[babyboy]

for babygirl in girls.keys():
    countgirls = countgirls + girls[babygirl]

print("How many boy names and girl names are described in the dataset?"+str(boy_names + girl_names))
print("How many boys and girls (actual babies!) are described in the dataset?:"+str(countboys+countgirls))
```

How often do the least common names occur? (Does your answer to this question bother you? Why?)

```python
import math;

more_common_name_boys_value = max(boys.values())
more_common_name_boys_key = ""
least_common_name_boys_value = min(boys.values())
least_common_name_boys_key = ""
more_common_name_girls_value = max(girls.values())
more_common_name_girls_key = ""
least_common_name_girls_value = min(girls.values())
least_common_name_girls_key = ""

for babyboy in boys.keys():
    if more_common_name_boys_value == boys[babyboy]:
        more_common_name_boys_key = babyboy
    if least_common_name_boys_value == boys[babyboy]:
        least_common_name_boys_key = babyboy

for babygirl in girls.keys():
    if more_common_name_girls_value == girls[babygirl]:
        more_common_name_girls_key = babygirl
    if least_common_name_girls_value == girls[babygirl]:
        least_common_name_girls_key = babygirl

print("What is the most common name for each gender in your data (i.e. 2021)?")
print("boys:"+more_common_name_boys_key+" "+str(more_common_name_boys_value))
print("girls:"+more_common_name_girls_key+" "+str(more_common_name_girls_value))

print("What is the least common name?")
print("boys:"+least_common_name_boys_key+" "+str(least_common_name_boys_value))
print("girls:"+least_common_name_girls_key+" "+str(least_common_name_girls_value))

print("How often do the least common names occur?")
print(str(least_common_name_boys_value)+"/"+str(countboys)) #previously obtained on #2
print(str(least_common_name_girls_value)+"/"+str(countgirls)) #previously obtained on #2
```

#6 Concentration in names
What percentage of boys have one of the 10 most popular boys names? What percentage of girls have one of the 10 most popular girls names?
Take the top 10% most popular boys names. (For instance: If there were 500 boys names, we're looking for the 50 most popular ones.) How many girls were given one of those names? Take the top 10% most popular girls names. How many boys were given one of those names?

```python
sorted_girls = dict(sorted(girls.items(), key=lambda item: item[1], reverse=True))
sorted_boys = dict(sorted(boys.items(), key=lambda item: item[1], reverse=True))
percent = 10
n = 1
print("Concentration in girls top 10:")
for current_name in sorted_girls:
    if n<=10:
        current_value = sorted_girls[current_name]
        current_percent = float(current_value/countgirls)*100
        print(str(n)+" "+current_name+" "+str(current_value)+"="+ str(current_percent)+"%")
    n = n+1

print("Concentration in boys top 10:")
n = 1
for current_name in sorted_boys:
    if n<=10:
        current_value = sorted_boys[current_name]
        current_percent = float(current_value/countboys)*100
        print(str(n)+" "+current_name+" "+str(current_value)+"="+ str(current_percent)+"%")
    n = n+1

#Take the top 10% most popular boys names. (For instance: If there were 500 boys names, we're looking for the 50 most popular ones.) 
# How many girls were given one of those names? Take the top 10% most popular girls names. How many boys were given one of those names?

print("Top "+str(percent)+"%"+"in boys from "+str(boy_names))
top_percent = boy_names/percent
n = 1
for current_name in sorted_boys:
    if n<=top_percent:
        print(str(n)+":"+current_name+" with a total of:"+str(sorted_boys[current_name]))
    n = n+1


print("Top "+str(percent)+"%"+"in boys from "+str(girl_names))
top_percent = girl_names/percent
n = 1
for current_name in sorted_girls:
    if n<=top_percent:
        print(str(n)+":"+current_name+" with a total of:"+str(sorted_girls[current_name]))
    n = n+1
```
