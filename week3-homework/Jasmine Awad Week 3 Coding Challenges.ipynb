```python
#What is the longest name in the dataset? How many 
#boys/girls names are exactly that length? What's 
#going on?

#The boys have one letter longer in their max names. 
#All of these names seem to be multiple names put together.

longest_names = []
max_length = 0
for name in girls:
    if len(name) > max_length:
        girl_longest_names = [name]
        max_length = len(name)
    elif len(name) == max_length:
        girl_longest_names.append(name)
for name in boys:
    if len(name) > max_length:
        boy_longest_names = [name]
        max_length = len(name)
    elif len(name) == max_length:
        boy_longest_names.append(name)

print(f"The longest girl names are {', '.join(girl_longest_names)} with {girls[girl_longest_names[0]]} babies.The longest boy names are {', '.join(boy_longest_names)} with {boys[boy_longest_names[0]]} babies.")
```

```python
#Initials to spreadsheets
```

```python
#Make a dictionary girl_initials that says, for each 
#letter of the alphabet (a-z), how many unique girl 
#names begin with that letter.
girl_initials = {}
for name in girls:
    first_letter = name[0]
    if first_letter in girl_initials:
        girl_initials[first_letter].add(name)
    else:
        girl_initials[first_letter] = set([name])

for letter in sorted(girl_initials.keys()):
    count = len(girl_initials[letter])
    print(f"{letter}: {count}")
```

```python
#Do the same for boy names to make boy_initials.
boy_initials = {}
for name in boys:
    first_letter = name[0]
    if first_letter in boy_initials:
        boy_initials[first_letter].add(name)
    else:
        boy_initials[first_letter] = set([name])

for letter in sorted(boy_initials.keys()):
    count = len(boy_initials[letter])
    print(f"{letter}: {count}")
```

```python
#Create a "tab separated values" (TSV) file that 
#reports the data in girl_initials. Be sure to include 
#a descriptive header columns! You will probably want 
#the file to end with .tsv so that your computer knows 
#it's a TSV file.

with open('girl_initials.tsv', 'w') as my_file: #tab separated values better than csv because commas are common
    print("girl_letter\tgirl_count", file = my_file) #prints the header for your file
    for letter in sorted(girl_initials.keys()):
        count = len(girl_initials[letter])
        print(f"{letter}\t{count}", file = my_file) #will print out letters and then number associated with letters into the file
```

```python
#Now do the same for boy_initials (be sure to save 
#it into a different file!)

with open('boy_initials.tsv', 'w') as my_file: #tab separated values better than csv because commas are common
    print("boy_letter\tboy_count", file = my_file) #prints the header for your file
    for letter in sorted(boy_initials.keys()):
        count = len(boy_initials[letter])
        print(f"{letter}\t{count}", file = my_file) #will print out letters and then number associated with letters into the file
```

```python
#What percentage of boys have one of the 10 most popular 
#boys names? What percentage of girls have one of the 10 
#most popular girls names?

total_boys = sum(boys.values())

top_10_boy_names = sorted(boys, key=boys.get, reverse=True)[:10]
top_10_boys = sum(boys[name] for name in top_10_boy_names)

boy_percentage = top_10_boys / total_boys * 100

total_girls = sum(girls.values())

top_10_girl_names = sorted(girls, key=girls.get, reverse=True)[:10]
top_10_girls = sum(girls[name] for name in top_10_girl_names)

girl_percentage = top_10_girls / total_girls * 100
print(f"{boy_percentage:.2f}% of boys have one of the 10 most popular boy names.{girl_percentage:.2f}% of girls have one of the 10 most popular girl names.")
```
