```python
print("There are " + str(len(boys)) + " boy names in the dataset.")
print("There are " + str(len(girls)) + " girl names in the dataset.")

# Now we count up boy babies and girl babies

count_boys = 0
for count_for_name in boys.values():
    count_boys = count_boys + count_for_name
print("There are " + str(count_boys) + " boys in the dataset.")

count_girls = 0
for count_for_name in girls.values():
    count_girls = count_girls + count_for_name
print("There are " + str(count_girls) + " girls in the dataset.")

# BONUS: Here's how to do it with `sum` (which replaces the for loop!)
print("BONUS...")
print("There are " + str(sum(boys.values())) + " boys in the dataset.")
print("There are " + str(sum(girls.values())) + " girls in the dataset.")
```

# 3

```python
most_common_boy_name_count = max(boys.values())
for name in boys.keys():
    if boys[name] == most_common_boy_name_count:
        print(f"Most common boy name is {name}, with {most_common_boy_name_count} babies.")

most_common_girl_name_count = max(girls.values())
for name in girls.keys():
    if girls[name] == most_common_girl_name_count:
        print(f"Most common girl name is {name}, with {most_common_girl_name_count} babies.")
```

```python
least_common_boy_name_count = min(boys.values())
for name in boys.keys():
    if boys[name] == least_common_boy_name_count:
        print(f"Least common boy name is {name}, with {least_common_boy_name_count} babies.")

least_common_girl_name_count = min(girls.values())
for name in girls.keys():
    if girls[name] == least_common_girl_name_count:
        print(f"Least common girl name is {name}, with {least_common_girl_name_count} babies.")
```

```python
# WOAH that's a lot of names with exactly 5 babies.
# I bet 5 is the minimum number of babies to have a name included in the dataset.
# Let's count how many of those 5-baby names there are.

boy_names_with_5_babies = 0
for name in boys.keys():
    if boys[name] == 5:
        boy_names_with_5_babies = boy_names_with_5_babies + 1
print(f"There are {boy_names_with_5_babies} names with 5 baby boys.")
        
girl_names_with_5_babies = 0
for name in girls.keys():
    if girls[name] == 5:
        girl_names_with_5_babies = girl_names_with_5_babies + 1
print(f"There are {girl_names_with_5_babies} names with 5 baby girls.")
```

```python
# Here's a funny approach: I'll make a new list of lengths,
# and then use `max` to figure out the longest.

name_lengths = []
for name in boys.keys():
    name_lengths.append(len(name))
for name in girls.keys():
    name_lengths.append(len(name))
max_name_lengths = max(name_lengths)
print(f"The longest name is {max_name_lengths} letters long.")
```

```python
girl_initials = {}
for name in girls.keys():
    initial = name[0]
    if initial in girl_initials:
        girl_initials[initial] = girl_initials[initial] + 1
    else:
        girl_initials[initial] = 1

boy_initials = {}
for name in boys.keys():
    initial = name[0]
    if initial in boy_initials:
        boy_initials[initial] = boy_initials[initial] + 1
    else:
        boy_initials[initial] = 1
```

```python
with open('girl_initials.tsv', 'w') as f:
    print("initial\tcount", file=f)
    for initial in girl_initials.keys():
        print(f"{initial}\t{girl_initials[initial]}", file=f)
```

```python
# Part 1

# The names are in the dictionary in order of popularity already.
# (Conveniently enough.)
# So we can just take the first 10.

# Actually, I'll do the same trick I do in part 2, with `sum`:

count_boys = sum(boys.values())
count_boys_top_10 = sum(list(boys.values())[:10])
# NOTE: I tried "sum(boys.values()[:10])" first, and it said:
#   'dict_values' object is not subscriptable
# That cued me in that boys.values() is a special kind of list, so I
# just wrapped it in `list` to make it a normal kind of list. Not at
# all obvious!
print(f"{count_boys_top_10 / count_boys :.1%} of boys have a top-10 name.")
# The ":.1%" thing above is a bit of magic formatting I looked up.

count_girls = sum(girls.values())
count_girls_top_10 = sum(list(girls.values())[:10])
print(f"{count_girls_top_10 / count_girls :.1%} of girls have a top-10 name.")
```

```python
# Huh, that's surprising!
# I'd heard that boys names were more concentrated, but that doesn't
# seem to be visible at the top-10 level.
```

```python
# Part 2

# First, figure out what top 10% means.

print(len(boys.values()) / 10)
print(len(girls.values()) / 10)
```

```python
# Now do the math.

girls_with_top_10pct_boys_names = 0
for name in list(boys.keys())[:1399]:
    # NOTE: We use .get with a default of 0 to deal with missing names
    girls_with_top_10pct_boys_names += girls.get(name, 0)
print(f"{girls_with_top_10pct_boys_names} girls were given a top-10% boys name")

boys_with_top_10pct_girls_names = 0
for name in list(girls.keys())[:1399]:
    # NOTE: We use .get with a default of 0 to deal with missing names
    boys_with_top_10pct_girls_names += boys.get(name, 0)
print(f"{boys_with_top_10pct_girls_names} boys were given a top-10% girls name")
```

```python
# Another surprise...
# The most popular girls names were more popular for boys than the
# most popular boys names were for girls!
```

# 7

```python
# Let's look at distribution of LAST letters
# (I'll call them "finals"; idk if that's conventional.)
# Also, I'll do an analysis of counting BABIES, not baby NAMES
# (cuz I think that will reveal more interesting patterns of popularity)
```

```python
girl_finals = {}
for name in girls.keys():
    final = name[-1]
    if not final in girl_finals:
        girl_finals[final] = 0
    girl_finals[final] += girls[name]

boy_finals = {}
for name in boys.keys():
    final = name[-1]
    if not final in boy_finals:
        boy_finals[final] = 0
    boy_finals[final] += boys[name]
```

```python
# To visualize, I'll cheat and use Pandas

import pandas as pd

df = pd.DataFrame({
    'girls': girl_finals,
    'boys': boy_finals,
}).sort_index()

df
```

```python
df.plot.bar()
```
